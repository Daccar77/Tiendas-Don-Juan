```python
data = {
    "Producto": ["Asistente virtual", "Mesa de comedor", "Juego de mesa", "Microondas", "Silla de oficina"],
    "Categoria": ["Electrónicos", "Muebles", "Juguetes", "Electrodomésticos", "Muebles"],
    "Precio": [164300.0, 192300.0, 209600.0, 757500.0, 335200.0],
    "Costo": [6900.0, 8400.0, 15900.0, 41000.0, 20200.0],
    "Fecha": ["16/01/2021", "18/05/2022", "15/03/2021", "03/05/2022", "07/11/2020"],
    "Vendedor": ["Pedro Gomez", "Beatriz Morales", "Juan Fernandez", "Juan Fernandez", "Maria Alfonso"],
    "Ciudad": ["Bogotá", "Medellín", "Cartagena", "Cali", "Medellín"],
    "Calificación": [4, 1, 1, 4, 5],
    "Método de pago": ["Tarjeta de crédito", "Tarjeta de crédito", "Tarjeta de crédito", "Nequi", "Nequi"],
    "Cuotas": [8, 4, 1, 1, 1],
    "Lat": [4.60971, 6.25184, 10.39972, 3.43722, 6.25184],
    "Lon": [-74.08175, -75.56359, -75.51444, -76.52250, -75.56359]
}
```

```python
df = pd.DataFrame(data)

df["Ganancia"] = df["Precio"] - df["Costo"]

ingreso_total = sum(df["Precio"])
costo_total = sum(df["Costo"])
ganancia_total = sum(df["Ganancia"])

print("Ingreso total:", ingreso_total)
print("Costo total:", costo_total)
print("Ganancia total:", ganancia_total)


```

```python
df[["Producto", "Precio", "Costo", "Ganancia"]]
```

# 2. Ventas por categoría

```python
import matplotlib.pyplot as plt

ventas_categoria = df.groupby("Categoria")["Precio"].sum().sort_values(ascending=False)
print("Ingreso total por categoria: ")
print(ingreso_total)
```

```python
plt.figure(figsize=(8, 5))
ventas_categoria.plot(kind="bar", color="skyblue")
plt.title("Ingresos total por categoria")
plt.xlabel("Categoria")
plt.ylabel("Ingreso")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = df["Calificación"].mean()
print(f"Calificación promedio de la tienda: {calificacion_promedio:.2f}")
```

```python

plt.figure(figsize=(6, 4))
df["Calificación"].value_counts().sort_index().plot(kind="bar", color="lightgreen")
plt.title("Distribución de Calificaciones de los Clientes")
plt.xlabel("Calificación")
plt.ylabel("Cantidad de Opiniones")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
ventas_productos = df.groupby("Producto")["Precio"].sum().sort_values(ascending=False)
print("Productos más vendidos:")
print(ventas_productos.head())
print("\nProductos menos vendidos:")
print(ventas_productos.tail())
```

```python
plt.figure(figsize=(8, 5))
ventas_productos.plot(kind="bar", color="lightblue")
plt.title("Productos más y menos vendidos")
plt.xlabel("Producto")
plt.ylabel("Ingreso")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
envio_ciudad = df.groupby("Ciudad")["Cuotas"].sum().sort_values(ascending=False)
print("Cantidad de productos enviados por ciudad:")
print(envio_ciudad)
```

```python
plt.figure(figsize=(7, 7))
envio_ciudad.plot(
    kind="pie",
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Set3.colors,  # Podés elegir otro colormap si querés
    ylabel=""  # Ocultamos el eje Y que sobra en pie charts
)
plt.title("Distribución de Productos Enviados por Ciudad")
plt.tight_layout()
plt.show()
```
